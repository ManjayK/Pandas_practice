```python
import pandas as pd 

dict1 = {'Name': ['Priyang Bhatt', 'Rashmin Panchal', 'Anil Rana', 'Rahul Patel'],
        'City': [' New York ', ' Los Angeles ', ' Chicago', ' Houston'],
        'State': ['NY', 'CA', 'IL', 'TX']}
data = pd.DataFrame(dict1)
```

```python
data
```

- str.strip() removes beginning and ending of string
- str.lstrip() removes spaces on the left side
-str.rstrip() removes spaces on the right side

```python
data['City'][0] # here we can see spaces in beginning and ending of the 'NewYork' which we can remove using strip()
```

```python
data['City']
```

```python
data['City_strip']=data['City'].str.strip()
```

```python
data
```

```python
data['City'][0]
```

```python
# now we will lstrip()

data['City_lstrip']=data['City'].str.lstrip()
```

```python
data
```

```python
# now we will use rstrip()

data['City_rstrip']=data['City'].str.rstrip()
```

# 4. str.split()

- It is commonly used function in pandas that allows to split a string into a list of sub-string based on specified separator.

```python
data['Name']
```

```python

# expand parameter is true which mean the result will be written as dataframe which one column for each sub-string.

data['Name'].str.split(' ',expand=True) 
```

```python
# now lets create a new column


data[['First_name','Last_name']]=data['Name'].str.split(' ',expand=True) 
```

```python
data
```


# 5. str.contains()

- the str.contains() function is used to check if a string contain a specified pattern or sub-string as TRUE OR FALSE where True indicates the pattern or sub-string is present and FALSE indicates that the pattern and sub-string is not present.

```python
data[data['Name'].str.contains("Bhatt")]
```

# 6.  str.replace()

- use to replace one or more sub-string in a string with another string. 

```python
data
```

```python
# lets replace the string 'NY' in state with New York

data['State_replaced']= data['State'].str.replace('NY','New York').str.replace('CA','California')
```

# 7. str.startswith() , str.endswith()

- It is used to check a string ends with a specified sub-string or not

```python
data
```

# 9. str.get()

- It allows to extract element of a string based on their index.

```python
data
```

```python
data['Name']
```

```python
# lets assume we want to extract first character from each name in this particular column.

data['Name'].str.get(0)
```

# 10. str.slice()

- It is used to extract sub-string from a pandas series of string.

```python
# lets extract last three characters of each name in the Name column

data['Name'].str.slice(-3)
```

```python
# lest extrace first three characters and concatinate it with "State" column
data['Name_state']=data['Name'].str.slice(0,3)+ ' ' + data['State'] 
```
